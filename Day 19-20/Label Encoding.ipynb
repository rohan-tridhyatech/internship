```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

#Lable Encoding for **Breast Cancer Dataset**

```python
cancer_data = pd.read_csv("/content/breast_cancer_data.csv")
```

```python
cancer_data.head()
```

```python
cancer_data["diagnosis"].value_counts()
```

```python
# load  LabelEncoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the label to the dataset
cancer_data["target"] = labels
```

```python
cancer_data.head()
```

**0 --> B(Benign)**<BR>
**1 --> M(Malignant)**

```python
cancer_data["target"].value_counts()
```
