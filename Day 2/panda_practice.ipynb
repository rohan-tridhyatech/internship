#Read CSV File

```python
# read csv file 
data = pd.read_csv("data.csv")
```

```python
# show last five line of data
data.tail()
```

```python
# number of rows and columns
data.shape
```

```python
# check null values
data.isnull().sum()
```

```python
# statistical analysis about data
data.describe()
```

```python
# check duplicates value
data.duplicated().value_counts()
```

```python
# drop duplicates data rows
data.drop_duplicates(inplace=True)
```
